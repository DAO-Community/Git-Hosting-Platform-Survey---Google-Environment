# Type Safety
Define constant objects to be expected in the data

```python
from enum import Enum


class Eyears_of_experience(Enum):
    less_than_two: str = "<2 years"
    two_to_four: str = "2 to 4 years"
    four_to_six: str = "4 to 6 years"
    eight_to_ten: str = "8 to 10 years"
    ten_plus: str = "10+ years"


class Egit_hosting_platform(Enum):
    github: str = "GitHub"
    gitlab: str = "GitLab"


class Ecareer_level(Enum):
    _1: str = "1"
    _2: str = "2"
    _3: str = "3"
    _4: str = "4"
    _5: str = "5"
    _6: str = "6"


class Edao_pillar(Enum):
    da: str = "Data Admin"
    de: str = "Data Engineering"
    dsa: str = "Data Science & Analytics"
    afc: str = "AFC"
    sc: str = "SolCon"


class Epast_next(Enum):
    past: str = "Worked with in PAST years"
    next: str = "Want to work with in NEXT years"


class Epast_next_all(Enum):
    past: str = Epast_next.past.value
    next: str = Epast_next.next.value
    past_next: str = Epast_next.past.value + ", " + Epast_next.next.value
```

```python
from typing import TypedDict

import polars as pl


class TDcolumns(TypedDict):
    timestamp: pl.DataType
    email_address: pl.DataType
    used_git_hosting_platform: pl.DataType
    current_git_hosting_platform: pl.DataType
    years_of_experience: pl.DataType
    past_next_github: pl.DataType
    past_next_gitlab: pl.DataType
    career_level: pl.DataType
    dao_pillar: pl.DataType


df_columns_dtypes_dict: TDcolumns = TDcolumns(
    timestamp=pl.Utf8,
    email_address=pl.Utf8,
    used_git_hosting_platform=pl.Utf8,
    current_git_hosting_platform=pl.Utf8,
    years_of_experience=pl.Utf8,
    past_next_github=pl.Utf8,
    past_next_gitlab=pl.Utf8,
    career_level=pl.Utf8,
    dao_pillar=pl.Utf8,
)
```
